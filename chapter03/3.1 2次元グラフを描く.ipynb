## 3.1.1 random graph

```python
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline

# create data
np.random.seed(1)
x = np.arange(10)
y = np.random.rand(10)

# drowing graph
plt.plot(x,y) #register graph
plt.show #drow graph
```

```python
# list 2-(1)
def f(x):
    return (x - 2) * x * (x + 2)
```

```python
# list2-(7)
#define function

def f2(x,w):
    return ( x - w) * x * (x + 2) #A

x = np.linspace(-3,3,100)#(b)

#drow graph
plt.plot(x,f2(x, 2),color='black', label=' $w = 2$') #(c)
plt.plot(x,f2(x, 1), color = 'cornflowerblue',
         label='$w=1$') #(D)

plt.legend(loc="upper left")
plt.ylim(-15,15)
plt.title('$f_2(x)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.grid(True)
plt.show()
```
